# Q1: Which branch has the highest total revenue?

```python
#group by Branch and sum of sales gives the total revenue
Revenues_Sum=df.groupby('Branch')["Sales"].sum()
Revenues_Sum
```

```python
#printimng the Branch with MaxTotalSales
CityName=Revenues_Sum.idxmax()
Revenue=Revenues_Sum.max()
print(f'{CityName} has highest total revenue of {Revenue}')
```

# Q2: Which product line brings the most revenue?
Group by Product line and sum Total.

```python
#Get sum of Sales of each Product Lines
productLines_Revenue=df.groupby("Product line")['Sales'].sum()
productLines_Revenue
```

```python
#Fetch the name of top product line and its total Revenue
topProductLine=productLines_Revenue.idxmax()
topRevenue=productLines_Revenue.max()
```

```python
print(f"Top product line {topProductLine} with {topRevenue} Revenue")
```

# Q3: Do members spend more per invoice than normal customers?
Group by Customer type and calculate average Total.


```python
#Calculating Avg sales of different Customer Types
customertypeSales=df.groupby('Customer type')['Sales'].mean()
customertypeSales
```

```python
BaseLine=customertypeSales['Normal']
```

```python
result = customertypeSales.apply(lambda x: 'High Spender' if x > BaseLine else 'Normal Spender')
result
```

## Q4: Which day of week has the highest total sales?
Group by DayOfWeek and sum Total.


```python
#Adding new column that shows Day of the week like 0=Mon, 6=Sun
df['DayOfWeek']=df['Date'].dt.dayofweek
```

```python
#Create a dictionary with keys as the Day number starting from 0 and Values as the actual day name 
day_names={0:'Monday',1:'Tuesday',2:'Wednesday',3:'Thursday',4:'Friday',5:'Saturday',6:'Sunday'}
```

```python
#Sum of Sales By grouping dayof the week
SalesofDays=df.groupby('DayOfWeek')['Sales'].sum().round(2)
print(SalesofDays)
```

```python
#Loop for Replacing Series index with day names
# Just the single line list comprehension does the below magic:

SalesofDays.index=[day_names[i] for i in SalesofDays.index]
print(SalesofDays)

# Equivalent for loop:
''' day_list=[]
for i in SalesofDays.index:
    day_list.append(day_names[i])  # Add day name to list
day_list
SalesofDays.index=day_list
SalesofDays '''
```

```python
HighestSalesDay=SalesofDays.idxmax()
HighestSalesDaySales=SalesofDays.max()
print(f'Highest sales Day is {HighestSalesDay} with {HighestSalesDaySales}' )
```

## Q5 (optional): Which city is most profitable (sum of gross income)?
